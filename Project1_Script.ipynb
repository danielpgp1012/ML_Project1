```python
#This version of cross validation returns the weights and the te
import random
def cross_validation_return(phi_2,y_tr):
    seeds = random.sample(range(1,50),4)
    degrees = 7
    k_fold = 4
    lambdas = np.logspace(-7, -5, 10)
    # split data in k fold
   
    # define lists to store the loss of training data and test data
    min_loss_te=100
    min_loss_te_list=np.zeros((len(seeds),))
    min_loss_tr_list=np.zeros((len(seeds),))
    min_lambda=np.zeros((len(seeds),))
    ws=np.zeros((len(seeds),phi_2.shape[1]))
    for index,seed in enumerate(seeds):
        k_indices = build_k_indices(y_tr, k_fold, seed)
        min_loss_te=100
        for lambda_ in lambdas:
            loss_tr_sum=0
            loss_te_sum=0
            for k in range(k_fold):
                loss_tr,loss_te,w = cross_validation(y_tr, phi_2, k_indices, k, lambda_, degrees)
                loss_tr_sum+=loss_tr
                loss_te_sum+=loss_te
            if (loss_te_sum/k_fold<min_loss):
                min_loss_te=loss_te_sum/k_fold
                min_loss_tr=loss_tr_sum/k_fold
                w_opt=w
                lambda_opt=lambda_
        ws[index,:]=w_opt
        min_lambda[index]=lambda_opt
        min_loss_te_list[index]=min_loss_te
        min_loss_tr_list[index]=min_loss_tr
   
    min_loss_te=np.mean(min_loss_te_list)
    min_loss_tr=np.mean(min_loss_tr_list)
    w_opt=np.mean(ws,axis=0)
    lambda_opt=np.mean(min_lambda)
    
    del min_loss_te_list
    del min_loss_tr_list
    del ws
    del min_lambda
    del lambdas
    del seeds
    
        
    # ***************************************************
    # INSERT YOUR CODE HERE
    # cross validation: TODO
    # ***************************************************    
    return min_loss_te,min_loss_tr,w_opt
    #plt.plot(range(2,degrees),min_losses_te)
    #plt.plot(range(2,degrees),min_losses_tr)
    #print(min_lambdas)
    #plt.hist(range(2,degrees),min_lambdas)
    
    
```
